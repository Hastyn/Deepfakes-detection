```python
### Download pre-trained model
import os
import tarfile
from coqui_stt_training.util.downloader import maybe_download

def download_pretrained_model():
    model_dir="english/"
    if not os.path.exists("english/coqui-yesno-checkpoints"):
        maybe_download("model.tar.gz", model_dir, "https://github.com/coqui-ai/STT-models/releases/download/english%2Fcoqui%2Fyesno-v0.0.1/coqui-yesno-checkpoints.tar.gz")
        print('\nNo extracted pre-trained model found. Extracting now...')
        tar = tarfile.open("english/model.tar.gz")
        tar.extractall("english/")
        tar.close()
    else:
        print('Found "english/coqui-yesno-checkpoints" - not extracting.')

# Download + extract pre-trained English model
download_pretrained_model()
```

```python
# Open the csv file and create two distinct good csvs for train and test

transcripts = {}

with open("marathi/marathi.csv") as csv:
  lines = csv.readlines()

  for line in lines:
    out = line.split('\t')
    transcripts[out[0] + ".wav"] = out[1][:-1]
```

```python
# Train files

train_set = os.listdir("marathi/wavs/train")
test_set = os.listdir("marathi/wavs/test")

with open("marathi/wavs/train/train.csv", "w") as out_train:
  out_train.write("wav_filename,wav_filesize,transcript\n")
  for f in train_set:
    try:
      line = f"{f},0,{transcripts[f]}"
      out_train.write(line + '\n')
    except:
      pass

with open("marathi/wavs/test/test.csv", "w") as out_test:
  out_test.write("wav_filename,wav_filesize,transcript\n")
  for f in test_set:
    try:
      line = f"{f},0,{transcripts[f]}"
      out_test.write(line + '\n')
    except:
      pass
```
